## Calculate the Geometric Mean and Standard Deviation

Calculating the geometric mean of the lognormal distribution:

\begin{equation}
\ln{d_g}=\frac{\sum{n_i}\ln{di}}{N}
\end{equation}

and the geometric standard deviation
\begin{equation}
\ln{\sigma_g}=\left(\frac{\sum{n_i}\left(\ln{d_i}-\ln{d_g}\right)^2}{N-1}\right)^{1/2}
\end{equation}
