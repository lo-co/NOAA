```python
import numpy as np
import math
import scipy.special as sp_spec
```

```python
def mie_abcd(m,x):
    '''
    Compute a matrix of Mie coefficients
    
    Parameters
    ----------
    m : complex
        Complex index of refraction
    x : float
        Size parameter where x = k0*a and k0 is the wavenumber in the ambient medium and a is the sphere radius
    '''
    nmax = round(2+x+r*x**(1/3))
    n = range(1, nmax)
    nu = (n+0.5)
    z = m*x
    sqx = math.sqrt(0.5*math.pi/x)
    sqz = math.sqrt(0.5*math.pi/z)
    
    # Bessel functions of the first kind
    bx = sp_spec.jv(nu, x)*sqx
    bz = sp_spec.jv(nu, z)*sqz
    
    # Bessel functions of the second kind
    yx = sp_spec.yv(nu, x)*sqx
    
    hx = bx+1j*yx
    
    b1x = [math.sin(x)/x, bx[0:nmax-2]]
    b1z = [math.sin(z)/z, bz[0:nmax-2]]
    y1x = [-math.cos(x)/x, yx[0:nmax-2]]
    h1x = b1x*1j*y1x
    ax = x*b1x-n*bx
    az = z*b1z-n*bz
    ahx = x*h1x-n*hx
    an = (m**2*bz*ax-bx*az)/(m**2*bz*ahx-hx*az)
    bn = (bz*ax-bx*az)/(bz*ahx-hx*az)
    cn = (bx*ahx-hx*ax)/(bz*ahx-hx*az)
    dn = m*(bx*ahx-hx*ax)/(m**2*bz*ahx-hx*az)
    return [an,bn,cn,dn]
```

```python
def Mie(m, x):
    '''
    % Computation of Mie Efficiencies for given
    % complex refractive-index ratio m=m'+im"
    % and size parameter x=k0*a, where k0= wave number in ambient
    % medium, a=sphere radius, using complex Mie Coefficients
    % an and bn for n=1 to nmax,
    % s. Bohren and Huffman (1983) BEWI:TDD122, p. 103,119-122,477.
    % Result: m', m", x, efficiencies for extinction (qext),
    % scattering (qsca), absorption (qabs), backscattering (qb),
    % asymmetry parameter (asy=<costeta>) and (qratio=qb/qsca).9
    % Uses the function "Mie_abcd" for an and bn, for n=1 to nmax.
    % C. Mätzler, May 2002.
    '''
    m = complex(m)
    if not(x): # To avoid a singularity at x=0
        print("hello")
        #result=[m.real m.imag 0 0 0 0 0 0 1.5]
    else:# This is the normal situation
        nmax=round(2+x+4*x**(1/3))-1
        n1=nmax-1
        n=np.linspace(0,nmax,nmax+1)
        #n=range(0,nmax)
        cn=2*n+1
        c1n=n*(n+2)/(n+1)
        c2n=cn/n/(n+1)
        x2=x*x
        
        g1 = np.zeros((4,nmax))
        
        f=mie_abcd(m,x)
        
        anp=f.real[0,]
        anpp=f.imag[0,]
        bnp=f.real[1,]
        bnpp=f.imag[1,]
        g1[0:3,nmax]=[0 0 0 0] # displaced numbers used for
        g1[1,0:n1]=anp[1:nmax-1] # asymmetry parameter, p. 120
        g1[2,0:n1]=anpp[1:nmax]
        g1[3,0:n1]=bnp[1:nmax]
        g1(4,0:n1)=bnpp[1:nmax]
        dn=cn*(anp+bnp)
        q=np.sum(dn)
        qext=2*q/x2
        en=cn*(anp*anp+anpp*anpp+bnp*bnp+bnpp*bnpp)
        q=np.sum(en)
        qsca=2*q/x2
        qabs=qext-qsca
        fn=(f[0,]-f[1,])*cn
        gn=(-1)**n
        f[2,]=fn*gn
        q=np.sum(f[2,])
        qb=q*np.transpose(q)/x2
        asy1=c1n*(anp*g1[0,]+anpp*g1[1,]+bnp*g1[2,]+bnpp*g1[3,])
        asy2=c2n*(anp*bnp+anpp*bnpp)
        asy=4/x2*np.sum(asy1+asy2)/qsca
        qratio=qb/qsca
        return [m.real m.imag x qext qsca qabs qb asy qratio]
```
